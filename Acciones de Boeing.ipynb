# Ejemplo de uso del sitio QUANDL para bajar datos que se actualizan periódicamente.

```python
# Localizar en el sitio http://quandl.com una base de datos a analizar (que sea gratuita!)
# En este caso, estamos localizando las acciones de la compañía de aeronáutica Boeing
# En la parte superior derecha aparecerá el código que hay que poner en el quandl.get
# Si no tienes la biblioteca quandl, para accesar directamente desde Python, puedes instalarla con
#          !pip3 install quandl
# o en una terminal teclea...
#          conda install quandl
#
# Si no tienes cuenta en quandl (se requiere, es gratuita!), entra al sitio y regístrate.


import quandl

quandl.ApiConfig.api_key = "consigue tu propia clave, es gratis..."
datos = quandl.get("EOD/BA")
```
