```python
ts =pd.Series(np.random.rand(100))
ts
```

```python
def matriz_de_recurrencia1(serie):
    """'serie' es una pandas.core.series.Series, indexada con 
    pandas.tseries.index.DatetimeIndex y valores numéricos."""
    tiempo=np.array(serie.index)
    n = tiempo.size
    xs = np.repeat(tiempo, repeats=n)
    ys = tiempo[np.newaxis].repeat(n, axis=0).flatten()
    mat = np.abs(serie[xs].values - serie[ys].values).reshape(n,n)
    plt.imshow(np.sqrt(mat), cmap=plt.cm.hot)
```

```python
matriz_de_recurrencia1(ts)
```

```python
def matriz_de_recurrencia2(serie):
    """'serie' es una pandas.core.series.Series."""
    xs, ys = np.meshgrid(serie, serie, indexing='ij')
    mat = np.abs(xs-ys)**2
    plt.imshow(mat, cmap=plt.cm.hot)
```

```python
matriz_de_recurrencia2()
```

```python
def matriz_de_recurrencia3(serie):
    """Identifica (los hace el mismo) el valor máximo con el mínimo de la serie."""
    xs, ys = np.meshgrid(serie, serie, indexing='ij')
    mat = np.abs(xs-ys)
    plt.imshow(mat, cmap=plt.cm.hot)
```
