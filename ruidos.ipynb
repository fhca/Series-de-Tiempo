```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=1000
seriep = np.random.rand(n)-.5
serie = np.cumsum(seriep)
```

```python
def H(v):
    suma = 0
    for i in range(n-v+1):
        suma += np.std(serie[i:i+v])
    promedio=suma/(n-v+1)  # div por el num total de ventanas
    return promedio
```

```python
plt.plot(serie)
plt.show()
# tamaños de ventana
vs = 2**np.arange(1, int(np.log2(n))+1)
# Calcula el exp de chapotin
e_chapotin_v = np.array([H(v) for v in vs])
e_chapotin_v.sort()
e_chapotin_v=e_chapotin_v[::-1]
plt.plot(e_chapotin_v)
plt.show()
# Ajuste de recta
r = np.polyfit(np.log2(vs), np.log2(e_chapotin_v), 1)
m, b = r  # pendiente y lado recto
plt.plot(np.log2(vs), np.log2(e_chapotin_v), '*')
plt.plot(m*np.log2(vs)+ b)
plt.show()
# Muestra curva ajustada
x = vs
k = 2**b
plt.plot(e_chapotin_v)
plt.plot(k * x**m)
plt.show()
print(m)
```
