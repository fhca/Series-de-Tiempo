```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
seriep = np.random.rand(1000)-.5
serie = np.cumsum(seriep)
plt.plot(serie)
```

```python
# tamaños de ventana
vs = np.array([2, 4, 8, 16, 32, 64, 128, 256, 512])

# h_i = desviación estándar de ventana i
# calcular H_v = promedio de las h_i para el tamaño de ventana v
```

```python
n=1000
v = 8
def H(v):
    suma = 0
    for i in range(n-v+1):
        suma += np.std(serie[i:i+v])
    promedio=suma/(n-v+1)  # div por el num total de ventanas
    return promedio
```

```python
e_chapotin_v = np.array([H(v) for v in vs])
```

```python
e_chapotin_v.sort()
e_chapotin_v=e_chapotin_v[::-1]
plt.plot(e_chapotin_v)
```

```python

plt.plot(np.log2(vs), np.log2(e_chapotin_v), '*')

# EL EXPONENTE DE CHAPOTIN SERÁ LA PENDIENTE DE LA RECTA (minimos cuadrados) 
# QUE MEJOR APROXIME ESTOS DATOS.
```

```python
r = np.polyfit(np.log2(vs), np.log2(e_chapotin_v), 1)
r
```
