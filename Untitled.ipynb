```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

JUEGO DEL CAOS

```python
p = np.array([[0,0], [1,0], [0.5, np.sqrt(2)]])
plt.scatter(*p.T)
x0 = np.array([.5,.5])
plt.scatter(x0[0],x0[1])
```

```python
r = np.random.randint(0, 3, 10000)
r
```

```python
vs = p[r]

ximenosuno = x0
xi=[]

for v in vs:
    xi.append( (v + ximenosuno)/2 )
    ximenosuno = xi[-1]  # x_{i-1} es el último agregado

x = np.array(xi).T # arreglo de x's, arreglo de y's

plt.scatter(*x, marker='.')
```

IFSs en un círculo

```python
a = np.linspace(0, 2*np.pi, 100, endpoint=False)
p = np.array([np.cos(a), np.sin(a)]).T
plt.scatter(*p.T)
plt.axis('Equal')
```

```python
def ifssuc(r):
    n = p.size[0]-1
    vs = p[n*r.astype(int)]
    ximenosuno = x0
    xi=[]
    for v in vs:
        xi.append( (v + ximenosuno)/2 )
        ximenosuno = xi[-1]  # x_{i-1} es el último agregado
    x = np.array(xi).T # arreglo de x's, arreglo de y's
    plt.scatter(*x, marker='.')
    plt.axis("Equal")
    
r = np.random.rand(10000)*2*np.pi
ifssuc(r)
```
