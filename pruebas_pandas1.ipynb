# Introducción a Pandas

Cargamos modulos y creamos un arreglo (de numpy) del 1...10000 y lo convertimos en bidimensional (de 100x100)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = np.arange(1,10001).reshape((100,100))
```

Mostramos el arreglo bidimensional (ojo: numpy tiene el tipo "matrix", que es diferente a estos arreglos bidimensionales, por ejemplo 'a * b' es multiplicación elemento por elemento 'a[i,j] * b[i,j]', si a y b son arreglos bidimensionales, y es multiplicación matricial, si son matrices.) 

```python
a
```

Convertimos a en una tabla (en Pandas se llaman "DataFrame"), nótense los nombres de renglones y de columnas.

```python
d = pd.DataFrame(a)
d
```

Los datos en un dataframe también pueden graficarse

```python
plt.figure()
plt.plot(d[0:100])
plt.show()
```

d2 tiene los elementos de a elevados al cuadrado. También lo graficamos. Nótese que son varias líneas de gráficos, una por columna.

```python
d2 = pd.DataFrame(a**2)
plt.plot(d2)
plt.show()
```

Los dataframe por default se "indexan" por columnas. Veamos la primera columna de d2:

```python
d2[0]
```

Veamos la primera columna de d

```python
d[0]
```

Nótese que en pandas se indexa por renglones. Este es el primer renglón del arreglo a.

```python
a[0]
```

Designemos el primer renglón como "header", es decir, nombres de columnas. El '[1]' del final significa que extaiga el 2o elemento de la lista de tablas. Así poblacion1 es una simple tabla.

```python
poblacion1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_largest_cities', header=0)[1]
poblacion1
```

Renombremos las columnas

```python
poblacion1.columns=['City', 'Nation', 'Image', 'City proper', 'Metropolitan area', 'Urban area']
poblacion1
```

Borremos el renglón 0, que tiene datos erróneos.

```python
poblacion1 = poblacion1[1:]
poblacion1
```

Borrando la columna Image, que contenía imágenes en la página web.

```python
del poblacion1['Image']
poblacion1
```
